# introduction

The dataset contains 45,000 records and 14 attributes related to personal and loan information. It includes details about individuals such as their age, gender, education level, income, employment experience, and homeownership status. Additionally, it provides information about the loans, including the loan amount, purpose, interest rate, percentage of income allocated to the loan, and the length of credit history. There are also columns for credit scores, previous loan defaults, and loan status, which indicates whether the loan was approved or not. This dataset is likely intended for analyzing factors influencing loan approvals or defaults.To analyze the dataset, we start by examining its structure and summary statistics using libraries like pandas and numpy. This includes identifying the data types, checking for missing values, and calculating metrics such as mean. For visualization, we use matplotlib and seaborn to create histograms, scatter plots, and bar charts.

1.WHAT IS THE AVERAGE INCOME OF INDIVIDUALS IN THE DATA?

```python
average_income=df["person_income"].mean()
print("average income:", average_income)
```

2.HOW DOES THE  AVERAGE LOAN AMOUNT VARY BY LOAN INTENT?

```python
avg_loan_by_intent=df.groupby("loan_intent")["loan_amnt"].mean()
print(avg_loan_by_intent)
```

3.WHICH IS THE HIGHEST INTERST RATE?

```python
df["loan_int_rate"].max()
```

4.HOW MANY RENT OWNERSHIP ARE THERE?

```python
ownership=df[df["person_home_ownership"]=="RENT"].count()
print(ownership)
df["person_home_ownership"].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.title("home ownership distribution")
plt.show()
```

5.IS THERE A CORRELATION BETWEEN CREDIT SCORE AND LOAN APPROVAL? 

```python
correlation=df[["credit_score","loan_status"]].corr()
print(correlation)
sns.heatmap(correlation)
plt.show()

```

6.WHICH LOAN AMOUNT IS HIGHER?

```python
df["loan_amnt"].max()
```

7.WHICH LOAN AMOUNT IS LOWER?

```python
df["loan_amnt"].min()
```

8.WHICH HOME OWNERSHIP TYPE IS MOST COMMON?

```python
home_ownership_counts=df["person_home_ownership"].value_counts()
print(home_ownership_counts)
plt.figure(figsize=(8,6))
home_ownership_counts.plot(kind="bar",edgecolor="black")
plt.title("Home ownership distribution")
plt.xlabel("home ownership")
plt.ylabel("count")
plt.show()
```

9.WHAT IS THE DISTRIBUTION OF LOAN AMOUNTS?

```python
loan_distribution=df["loan_amnt"].describe()
print(loan_distribution)
plt.figure(figsize=(10,6))
plt.hist(df["loan_amnt"],bins=10,edgecolor="black")
plt.title("distribution of loan amnt")
plt.xlabel("loan amount")
plt.ylabel("frequency")
plt.show()
```

10.GRAPHICALLY REPRESENT THE PERSON EDUCATION?

```python
cn=df["person_education"].value_counts()
print(cn)
plt.figure(figsize=(8,6))
cn.plot(kind="bar",edgecolor="black")
plt.title("Home ownership distribution")
plt.xlabel("home ownership")
plt.ylabel("count")
plt.show()
```

11.CALCULATE THE SUM OF LOAN PROVIDED FOR MALE?

```python
male_loan_sum=df.groupby("person_gender")["loan_amnt"].sum().loc["male"]
print(male_loan_sum)
```

12.CALCULATE MALE AND FEMALE COUNT?

```python
gender_counts=pd.pivot_table(df,index="person_gender",aggfunc="size")
print(gender_counts)
plt.figure(figsize=(8,8))
plt.pie(gender_counts,labels=gender_counts.index,autopct="%1.1f%%")
plt.title("male & female")
plt.show()

```

13.CALCULATE THE SUM OF LOAN PROVIDED FOR MALE?

```python
male_loan_sum=df.groupby("person_gender")["loan_amnt"].sum().loc["male"]
print(male_loan_sum)
```

14.HOW MANY PERSONS HAD THE INCOME ABOVE ONE LAKH?

```python
high_income_count=(df["person_income"]>100000).sum()
print("number of persons with income above one lakh:",high_income_count)
```

15.FIND THE RELATION BETWEEN LOAN INTEREST RATE AND LOAN PERCENTAGE INCOME?

```python
plt.figure(figsize=(8,6))
plt.scatter(df["loan_int_rate"],df["loan_percent_income"])
plt.xlabel("loan_int_rate")
plt.ylabel("loan_percent_income")
plt.title("relationship between loan interest rate and loan percent income")
plt.show()
```
